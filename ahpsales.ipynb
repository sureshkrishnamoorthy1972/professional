```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
division=['RUMICARE','AVICARE','AQUACARE','OMNICARE','GOVT']
target=[147.60,65.17,3.75,58.20,2.30]
```

```python
xpos = np.arange(len(division))
xpos
```

```python
plt.bar(xpos,target, label="target")
plt.xticks(xpos, division)
plt.ylabel("target(crs)")
plt.title('AHP Sales Target FY21')
plt.legend()
```

```python
PY=[131.46,57.82,3.01,45.60,2.20]
```

```python
plt.bar(xpos-0.2,target, width=0.4, label="Target")
plt.bar(xpos+0.2,PY, width=0.4,label="PY")
plt.xticks(xpos,division)
plt.ylabel("target(crs)")
plt.title('AHP Sales Targets for FY21')
plt.legend()
```
